```python
import itertools
import numpy as np
import matplotlib

import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

class_names = ["bicolor_no_madura", "madura", "no_madura", "otros", "podrida"]

y_true = []
y_pred = []

for i in range(0, 310):
    y_true.append("bicolor_no_madura")
    y_pred.append("bicolor_no_madura")
for i in range(0, 9):
    y_true.append("bicolor_no_madura")
    y_pred.append("madura")
for i in range(0, 11):
    y_true.append("bicolor_no_madura")
    y_pred.append("no_madura")
for i in range(0, 8):
    y_true.append("bicolor_no_madura")
    y_pred.append("otros")
for i in range(0, 0):
    y_true.append("bicolor_no_madura")
    y_pred.append("podrida")

for i in range(0, 1):
    y_true.append("madura")
    y_pred.append("bicolor_no_madura")
for i in range(0, 385):
    y_true.append("madura")
    y_pred.append("madura")
for i in range(0, 0):
    y_true.append("madura")
    y_pred.append("no_madura")
for i in range(0, 7):
    y_true.append("madura")
    y_pred.append("otros")
for i in range(0, 25):
    y_true.append("madura")
    y_pred.append("podrida")

for i in range(0, 28):
    y_true.append("no_madura")
    y_pred.append("bicolor_no_madura")
for i in range(0, 0):
    y_true.append("no_madura")
    y_pred.append("madura")
for i in range(0, 312):
    y_true.append("no_madura")
    y_pred.append("no_madura")
for i in range(0, 11):
    y_true.append("no_madura")
    y_pred.append("otros")
for i in range(0, 0):
    y_true.append("no_madura")
    y_pred.append("podrida")

for i in range(0, 0):
    y_true.append("otros")
    y_pred.append("bicolor_no_madura")
for i in range(0, 0):
    y_true.append("otros")
    y_pred.append("madura")
for i in range(0, 9):
    y_true.append("otros")
    y_pred.append("no_madura")
for i in range(0, 554):
    y_true.append("otros")
    y_pred.append("otros")
for i in range(0, 10):
    y_true.append("otros")
    y_pred.append("podrida")
    
for i in range(0, 0):
    y_true.append("podrida")
    y_pred.append("bicolor_no_madura")
for i in range(0, 11):
    y_true.append("podrida")
    y_pred.append("madura")
for i in range(0, 0):
    y_true.append("podrida")
    y_pred.append("no_madura")
for i in range(0, 0):
    y_true.append("podrida")
    y_pred.append("otros")
for i in range(0, 369):
    y_true.append("podrida")
    y_pred.append("podrida")

    
    
    
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print('Pnasnet_Large, normalizada')
    else:
        print('Pnasnet_Large, sin normalizar')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()

# Compute confusion matrix
cnf_matrix = confusion_matrix(y_true, y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes= ["bicolor_no_madura", "madura", "no_madura", "otros", "podrida"],
                      title='Pnasnet_Large, sin normalizar')

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names, normalize=True,
                      title='Pnasnet_Large, normalizada')

plt.show()
```

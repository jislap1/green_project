```python
import itertools
import numpy as np
import matplotlib

import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

class_names = ["cesped_bueno", "otros", "regar_cesped", "replantar_cesped", "sembrar_cesped"]

y_true = []
y_pred = []

for i in range(0, 346):
    y_true.append("cesped_bueno")
    y_pred.append("cesped_bueno")
for i in range(0, 81):
    y_true.append("cesped_bueno")
    y_pred.append("otros")
for i in range(0, 136):
    y_true.append("cesped_bueno")
    y_pred.append("regar_cesped")
for i in range(0, 13):
    y_true.append("cesped_bueno")
    y_pred.append("replantar_cesped")
for i in range(0, 9):
    y_true.append("cesped_bueno")
    y_pred.append("sembrar_cesped")

for i in range(0, 3):
    y_true.append("otros")
    y_pred.append("cesped_bueno")
for i in range(0, 826):
    y_true.append("otros")
    y_pred.append("otros")
for i in range(0, 17):
    y_true.append("otros")
    y_pred.append("regar_cesped")
for i in range(0, 24):
    y_true.append("otros")
    y_pred.append("replantar_cesped")
for i in range(0, 16):
    y_true.append("otros")
    y_pred.append("sembrar_cesped")

for i in range(0, 54):
    y_true.append("regar_cesped")
    y_pred.append("cesped_bueno")
for i in range(0, 135):
    y_true.append("regar_cesped")
    y_pred.append("otros")
for i in range(0, 425):
    y_true.append("regar_cesped")
    y_pred.append("regar_cesped")
for i in range(0, 172):
    y_true.append("regar_cesped")
    y_pred.append("replantar_cesped")
for i in range(0, 1):
    y_true.append("regar_cesped")
    y_pred.append("sembrar_cesped")

for i in range(0, 3):
    y_true.append("replantar_cesped")
    y_pred.append("cesped_bueno")
for i in range(0, 106):
    y_true.append("replantar_cesped")
    y_pred.append("otros")
for i in range(0, 81):
    y_true.append("replantar_cesped")
    y_pred.append("regar_cesped")
for i in range(0, 285):
    y_true.append("replantar_cesped")
    y_pred.append("replantar_cesped")
for i in range(0, 23):
    y_true.append("replantar_cesped")
    y_pred.append("sembrar_cesped")
    
for i in range(0, 1):
    y_true.append("sembrar_cesped")
    y_pred.append("cesped_bueno")
for i in range(0, 79):
    y_true.append("sembrar_cesped")
    y_pred.append("otros")
for i in range(0, 3):
    y_true.append("sembrar_cesped")
    y_pred.append("regar_cesped")
for i in range(0, 43):
    y_true.append("sembrar_cesped")
    y_pred.append("replantar_cesped")
for i in range(0, 345):
    y_true.append("sembrar_cesped")
    y_pred.append("sembrar_cesped")

    
    
    
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print('ResNet_v2, normalizada')
    else:
        print('ResNet_v2, sin normalizar')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()

# Compute confusion matrix
cnf_matrix = confusion_matrix(y_true, y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes= ["cesped_bueno", "otros", "regar_cesped", "replantar_cesped", "sembrar_cesped"],
                      title='ResNet_v2, sin normalizar')

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=class_names, normalize=True,
                      title='ResNet_v2, normalizada')

plt.show()
```
